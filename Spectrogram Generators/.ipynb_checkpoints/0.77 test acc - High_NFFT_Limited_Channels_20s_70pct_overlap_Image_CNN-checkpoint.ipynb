```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import math


from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Dropout, Bidirectional
from tensorflow.keras.layers import Input, Flatten
from tensorflow.keras.layers import Conv2D, Conv1D, MaxPooling2D, MaxPooling1D, AveragePooling2D, AveragePooling1D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dense, Activation, Permute, Dropout
from tensorflow.keras.models import Sequential

```

```python
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout
from tensorflow.keras.models import Sequential

def image_cnn_model(img_width, img_height):
    model = Sequential()
    model.add(Conv2D(8,(3,3),activation='relu',input_shape=(img_width,img_height,3)))
    model.add(MaxPool2D((2,2)))
    model.add(Dropout(0.15))

    model.add(Conv2D(8,(6,6),activation='relu'))
    #model.add(MaxPool2D((2,2)))
    model.add(Dropout(0.15))
    
    #model.add(Conv2D(64,(3,3),activation='relu'))
    #model.add(MaxPool2D((2,2)))
    #model.add(Dropout(0.15))
    model.add(Flatten())
    model.add(Dense(4,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    
    return model
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
import tensorflow.keras.backend as K


model = image_cnn_model(img_width, img_height)
model.summary()

def rmse (y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred -y_true)))
metrics = ['accuracy',  rmse,
            tf.keras.metrics.MeanSquaredError(),
            tf.keras.metrics.AUC(),
            'mse', 'mae']

image_cnn_opt_adam = tf.keras.optimizers.Adam(lr=0.00008, 
                                beta_1=0.99,
                                beta_2=0.95,
                                epsilon=1e-07)

model.compile(loss='binary_crossentropy', optimizer=image_cnn_opt_adam, metrics=metrics)

filepath = 'img_cnn_weights_stw.hdf5'

checkpoint = ModelCheckpoint(filepath, monitor='val_acc', 
                             verbose=0, save_best_only=True, save_weights_only=True, mode='auto',period=1)

```

```python
from sklearn.model_selection import train_test_split
import ctypes
import os, shutil
import time
from shutil import copyfile

from random import randrange


def clear_directory_contents(directory_location):
    if os.path.exists(directory_location):
        for filename in os.listdir(directory_location):
            file_path = os.path.join(directory_location, filename)
            try:
                if os.path.isfile(file_path) or os.path.islink(file_path):
                    os.unlink(file_path)
                elif os.path.isdir(file_path):
                    shutil.rmtree(file_path)
            except Exception as e:
                print('Failed to delete %s. Reason: %s' % (file_path, e))



# use copyfile to populate the train, test and validation directories :)
def create_symlinks(group_directories, group_file_name, output_directory):

        for j, directory in enumerate(group_directories):
            cwd = str(os.getcwd()) + '/'
            group = directory.split('/')[-1][:2]  #hc or sz
            src = cwd  + directory
            dst = cwd + output_directory + '/' + group + '/' + group_file_name + str(j)
            if  'COLAB_GPU' in os.environ : 
                src = directory.replace('/content/', '/content/drive/My Drive/')
                dst = output_directory + '/' + group + '/' + group_file_name + str(j)
            if j == 0:
                print('Declared source and destination addresses: ')
                print('src: ', src)
                print('dst: ', dst)

            for img_file in os.listdir(src):
                if  'COLAB_GPU' not in os.environ : 
                    os.symlink(src + '/' + img_file, dst + '_' + img_file, target_is_directory=False)
                else: #symlinks are not permitted in colab
                    copyfile(src + '/' + img_file,   dst + '_' + img_file,)
          

    
def test_sz_only(checkpoint_file_path, image_files_sub_dir, img_width, img_height,
                 batch_size=20, iterations=5, optimizer='Adam', metrics=['accuracy'],
                 loss='binary_crossentropy'):
    # List of raw patient file IDs that should be skipped based on categorization as outliers
    ignore_list = ['h09', 'h10', 's10', 's11', 's12']

    # generate the list of subject ids for file retrieval. File system and ignore list have different formats
    hc_subject_ids = ['hc' + str(i) for i in range(14) if "h{:02}".format(i) not in ignore_list] 
    sz_subject_ids = ['sz' + str(i) for i in range(14) if "s{:02}".format(i) not in ignore_list] 
    
    all_subject_ids = np.concatenate([hc_subject_ids, sz_subject_ids], axis=0)
    
    
    history_list = []
    test_metrics_list = []
    for i in range(iterations):
        rand_seed = 1
        #train_hc, test_hc = train_test_split(hc_subject_ids, test_size=0.2, random_state=rand_seed)
        #train_sz, test_sz = train_test_split(sz_subject_ids, test_size=0.2, random_state=rand_seed)
        #train_ids = np.concatenate([train_hc, train_sz])
        #test_ids = np.concatenate([test_hc, test_sz])
        #ids = {'train': train_ids, 'test': test_ids}

        rand_seed = 1 #randrange(100)
        print('\n\nBeginning iteration {} of {} '.format(i+1,iterations))
        print('\nRandom seed for participant selection: ', rand_seed )

        # copy the original list of  patient ids to allow for direct modifications
        iter_subject_ids = np.array(all_subject_ids.copy()).tolist()
        
        # get hc and sz subject list for this iteration and split them 
        # (to ensure that train and val sets have a balanced number of sz and hc participants)
        iter_hc_subject_ids = [s for s in iter_subject_ids if s.startswith('hc')]
        iter_sz_subject_ids = [s for s in iter_subject_ids if s.startswith('sz')]

  

        # split into training, test and validation
        train_hc, validate_hc = train_test_split(iter_hc_subject_ids, test_size=0.2, random_state=rand_seed)
        train_sz, validate_sz = train_test_split(iter_sz_subject_ids, test_size=0.2, random_state=rand_seed)
        train_hc, test_hc = train_test_split(train_hc, test_size=0.25, random_state=rand_seed)
        train_sz, test_sz = train_test_split(train_sz, test_size=0.25, random_state=rand_seed)
        iter_train = np.concatenate([train_hc, train_sz])
        iter_validate = np.concatenate([validate_hc, validate_sz])
        iter_test = np.concatenate([test_hc, test_sz]) # testing only sz patients"""
        
        print('\nSubjects assigned to groups using sklearn.model_selection.train_test_split')
        print('Training group: ', ", ".join(iter_train))
        print('Validation group: ', ", ".join(iter_validate), "\n")
        print('Test group: ', ", ".join(iter_test), "\n")
        

        train_directories = []
        validate_directories = []
        test_directories = []

        # create the addresses that will be used by image generator
        train_input_directory = image_files_sub_dir + 'train'
        validate_input_directory = image_files_sub_dir + 'validate'
        test_input_directory = image_files_sub_dir + 'test'

        # reconstruct the directories for the subjects in each set
        for sid in iter_train:
            subject_directory = image_files_sub_dir + sid[:2] + '/' + sid # eg: data_by_subject/hc/hc01
            train_directories.append(subject_directory)
        for sid in iter_validate:
            subject_directory = image_files_sub_dir + sid[:2] + '/' + sid # eg: data_by_subject/hc/hc01
            validate_directories.append(subject_directory)
        for sid in iter_test:
            subject_directory = image_files_sub_dir + sid[:2] + '/' + sid # eg: data_by_subject/hc/hc01
            test_directories.append(subject_directory)
        print(validate_directories)
        print(train_directories)
        #print(a)
        #confirm presence

        #if os.path.isdir(train_input_directory):
        #    #print('\nChecking number of files in train input directory BEFORE clearing: ',
        #    #    len([name for name in os.listdir(train_input_directory) if os.path.isfile(name)]))
            
        # if the directories already exist, empty them
        clear_directory_contents(train_input_directory )
        clear_directory_contents(validate_input_directory )
        clear_directory_contents(test_input_directory )
        time.sleep(20)    # pause 20 seconds to avoid PermissionError when creating directories immediately after clearing files from last iteration

        #confirm clear
        #if os.path.isdir(train_input_directory):
        #    #print('\nConfirming number of files in train input directory AFTER clearing: ',
        #    #    len([name for name in os.listdir(train_input_directory) if os.path.isfile(name)]))

        # create train_symlinks and validate_symlinks directories
        #if not os.path.exists(train_input_directory + '/hc'):
        os.makedirs(train_input_directory + '/hc')
        os.makedirs(train_input_directory + '/sz')
        os.makedirs(validate_input_directory + '/hc')
        os.makedirs(validate_input_directory + '/sz')
        os.makedirs(test_input_directory + '/sz')
        os.makedirs(test_input_directory + '/hc')



        # iterate ove the reconstructed directories for train and validate and create symlinks for each
        print('\nCreating symlinks')
        #kdll = ctypes.windll.LoadLibrary("kernel32.dll")
        create_symlinks(train_directories, 'train_file_', train_input_directory)
        create_symlinks(validate_directories, 'validate_file_', validate_input_directory)
        create_symlinks(test_directories, 'test_file_', test_input_directory)



        print('\nCreating data generators')
        train_datagen = ImageDataGenerator(rescale=1./255)
        validation_datagen = ImageDataGenerator(rescale=1./255)
        print('train_input_directory: ', train_input_directory)

        train_generator = train_datagen.flow_from_directory(train_input_directory,target_size=(img_width,img_height),
                                                    batch_size=batch_size,class_mode='binary',interpolation='bilinear')
        validation_generator = validation_datagen.flow_from_directory(validate_input_directory,target_size=(img_width,img_height),
                                                    batch_size=batch_size,class_mode='binary',interpolation='bilinear')


        #declare a new model within the loop to avoid possibility of weights being copied over
        model = image_cnn_model(img_width, img_height)
        model.summary()
        model.compile(loss='binary_crossentropy', optimizer=image_cnn_opt_adam, metrics=metrics)


        #subect_data = data_reader.chunk_list(subject_data, time_window)        
        print('\nBeginning model training...')
        filepath = checkpoint_file_path + str(i)

        checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', 
                                     verbose=0, save_best_only=True, save_weights_only=True, mode='auto',period=1)

        history = model.fit_generator(train_generator, steps_per_epoch=math.floor(train_generator.samples // batch_size), 
                                                    epochs=120, validation_data=validation_generator,
                                                    validation_steps=math.floor(validation_generator.samples // batch_size),  
                                                                                callbacks=[checkpoint])

        history_list.append(history)
        model_builder.print_model_metrics_multiple_iterations([history], acc_key)

        #### get test metrics
        #test_subject_directory = image_files_sub_dir + subject_id[:2] + '/' + subject_id # eg: data_by_subject/hc/hc01
        #print('test subject directory: ' + test_subject_directory)
        test_datagen = ImageDataGenerator(rescale=1./255)
        test_generator = test_datagen.flow_from_directory(
          test_input_directory,
          target_size=(img_width,img_height),
          batch_size=batch_size,
          class_mode='binary')
        
        probabilities = model.predict(test_generator, steps=test_generator.samples//batch_size)
        predicted_class_indices=np.argmax(probabilities,axis=1)

        from sklearn.metrics import confusion_matrix
        test_labels = (validation_generator.class_indices)
        labels2 = dict((v,k) for k,v in test_labels.items())
        #predictions = [test_labels[k] for k in predicted_class_indices]
        #print(predicted_class_indices)
        print('predictions : ', ", ".join([str(p) for p in probabilities]))
        #print(test_labels)
        #print (predictions)

        #y_true = np.array([0] * 1000 + [1] * 1000)

        #####y_pred = probabilities > 0.5
        #####confusion_matrix(test_labels, y_pred)
        #####Found input variables with inconsistent numbers of samples: [2, 140]        

        test_metrics = model.evaluate(x=test_generator, batch_size=batch_size)
        print('Test Metrics: ', test_metrics)
        metric_titles = ['Loss', 'Accuracy', 'RMSE', 'Mean Squared Error', 'AUC', 'MSE', 'MAE']
        metrics_string = "\n".join([(metric_titles[i]+": " + str(metric)) for i,metric in enumerate(test_metrics) ])
        print(metrics_string)
        ######print('Test accuracy: {0:.2f}%   | Test loss: {1:.5f} | Test RMSE: {2:.5f}'
        ######     .format(test_acc * 100, test_loss, test_rmse))
        #test_metrics = {'test_loss': test_loss, 
        #                'test_acc': test_acc,
        #                'test_rmse': test_rmse}
        test_metrics_list.append(test_metrics)
        reload(tf)
        reload(K)


    return history_list, test_metrics_list


# def plot_loo(history_list)     
history_list_sz_test, test_metrics_list_sz_test = test_sz_only('image_model_checkpoint', 
                                                                image_files_sub_dir=output_directory, 
                                                                img_width=img_width, img_height=img_height, 
                                                                metrics=metrics, optimizer=image_cnn_opt_adam, 
                                                                loss='binary_crossentropy',  iterations=5) #output_directory = 'data_by_subject/'
```
